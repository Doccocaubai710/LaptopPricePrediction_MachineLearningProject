## Splitting the dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 4)
```

## Train models

```python
# import models
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor, ExtraTreesRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

### Linear regression

```python
step1 = ColumnTransformer(transformers = [('col_tnf', OneHotEncoder(sparse = False, drop = 'first'), 
                                           [0, 1, 7, 12, 13])], remainder = 'passthrough')

step2 = LinearRegression()

pipe = Pipeline([('step1', step1), ('step2', step2)])

pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
print('R2 Score: ', r2_score(y_test, y_pred))
print('MAE Score: ', mean_absolute_error(y_test, y_pred))
```

## Ridge regression

```python
step1 = ColumnTransformer(transformers = [('col_tnf', OneHotEncoder(sparse= False, drop = 'first'), 
                                           [0, 1, 7, 12, 13])], remainder = 'passthrough')

step2 = Ridge(alpha = 10)

#Define the alpha value to test
alpha_values=[0.1,0.5,2,5,10,20,50,100]
r2_scores=[]
mae_scores=[]
pipe=Pipeline([('step1',step1),('step2',step2)])
for alpha in alpha_values:
    step2=Ridge(alpha=alpha)
    

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)

    # Calculate R2 score and mean absolute error
    r2_scores.append(r2_score(y_test, y_pred))
    mae_scores.append(mean_absolute_error(y_test, y_pred))

#Visualize the scores:
plt.figure(figsize=(10,6))
plt.plot(alpha_values,r2_scores,'bo-',label='R2_score')
plt.plot(alpha_values,mae_scores,'bo-',label='MAE score')
plt.xlabel('Alpha')
plt.ylabel('Score')
plt.grid(True)
plt.title("Visualize the score with alpha changing")
plt.show()

```

## KNN

```python
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV
step1 = ColumnTransformer(transformers = [('col_tnf', OneHotEncoder(sparse = False, drop = 'first'), 
                                           [0, 1, 7, 12, 13])], remainder = 'passthrough')
# Define the hyperparameters to be tuned
param_grid = {'step2__n_neighbors': [5, 10, 15, 20]}
# Define the evaluation metric
scoring = {'R2': make_scorer(r2_score), 'MAE': 'neg_mean_absolute_error'}
step2=KNeighborsRegressor()
pipe = Pipeline([('step1', step1), ('step2', step2)])


# Create the GridSearchCV object
grid_search = GridSearchCV(pipe, param_grid, scoring=scoring, refit='R2', cv=5)

# Fit the model to the training data
grid_search.fit(X_train, y_train)

# Get the best hyperparameters and the best estimator
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Make predictions on the test data using the best model
y_pred = best_model.predict(X_test)

# Calculate R2 score and mean absolute error
r2score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

# Print the evaluation metrics
print('Best Hyperparameters:', best_params)
print('R2 score:', r2score)
print('MAE:', mae)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of values for n_neighbors
n_neighbors_values = [5, 10, 15, 20]

# Initialize empty lists to store the evaluation scores
r2_scores = []
mae_scores = []

# Iterate over each value of n_neighbors
for n in n_neighbors_values:
    # Create and fit the KNeighborsRegressor model
    step2 = KNeighborsRegressor(n_neighbors=n)
    pipe=Pipeline([('step1',step1),('step2',step2)])
    pipe.fit(X_train, y_train)
    
    # Make predictions on the test data
    y_pred = pipe.predict(X_test)
    
    # Calculate R2 score and mean absolute error
    r2score = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    
    # Append the scores to the respective lists
    r2_scores.append(r2score)
    mae_scores.append(mae)

# Plotting the scores
plt.plot(n_neighbors_values, r2_scores, label='R2 Score')
plt.plot(n_neighbors_values, mae_scores, label='MAE')
plt.xlabel('n_neighbors')
plt.ylabel('Score')
plt.title('Evaluation Scores for different n_neighbors')
plt.legend()
plt.show()
```

## SVM

```python
step1 = ColumnTransformer(transformers = [('col_tnf', OneHotEncoder(sparse= False, drop = 'first'), 
                                           [0, 1, 7, 12, 13])], remainder = 'passthrough')

step2 = SVR(kernel = 'rbf', C = 100000, epsilon = 0.1)

# Define a range of C values to test
C_values = [0.1, 1, 10, 100, 1000,10000,100000]

# Define epsilon values
epsilon_values = [0.1, 0.5]

# Initialize lists to store the evaluation scores
r2_scores = {}
mae_scores = {}

# Fit the SVR model with different C values and evaluate the performance
for epsilon in epsilon_values:
    r2_scores[epsilon] = []
    mae_scores[epsilon] = []
    for C in C_values:
        step2 = SVR(kernel='rbf', C=C, epsilon=epsilon)
        pipe=Pipeline([('step1',step1),('step2',step2)])
        pipe.fit(X_train, y_train)
        y_pred = pipe.predict(X_test)
        r2_scores[epsilon].append(r2_score(y_test, y_pred))
        mae_scores[epsilon].append(mean_absolute_error(y_test, y_pred))

# Plot the evaluation scores against different C values for each epsilon
plt.figure(figsize=(12, 5))

# Plot for epsilon = 0.1
plt.subplot(1, 2, 1)
plt.plot(C_values, r2_scores[0.1], 'o-', label='R2 Score')
plt.plot(C_values, mae_scores[0.1], 'o-', label='Mean Absolute Error')
plt.xlabel('C')
plt.ylabel('Score')
plt.title('SVR Performance with Epsilon = 0.1')
plt.legend()
plt.grid(True)

# Plot for epsilon = 0.5
plt.subplot(1, 2, 2)
plt.plot(C_values, r2_scores[0.5], 'o-', label='R2 Score')
plt.plot(C_values, mae_scores[0.5], 'o-', label='Mean Absolute Error')
plt.xlabel('C')
plt.ylabel('Score')
plt.title('SVR Performance with Epsilon = 0.5')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
