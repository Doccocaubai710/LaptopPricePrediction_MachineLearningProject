WORK WITH MEMORY



```python
df['Memory'].value_counts()
```

```python
#Split the memory based on SSD,HDD,TB, Hybrid and Flash Storage
df['Memory']=df['Memory'].astype(str).replace('\.0','',regex=True)
df['Memory']=df['Memory'].str.replace('GB','')
df['Memory']=df['Memory'].str.replace('TB','000')
new=df['Memory'].str.split('+',n=1,expand=True)
#n=1 mean that only one split perform, the splitting occurs at the first occurence of the '+' character
#expand=True : When expand=True, the split element are returned as separate columns.In this case, there will be two columns, one for those before
#the + and one for those after the +

df['first']=new[0]
df['first']=df['first'].str.strip()


df['Layer1HDD']=df['first'].apply(lambda x:1 if 'HDD' in x else 0)
df['Layer1SSD']=df['first'].apply(lambda x:1 if 'SSD' in x else 0)
df['Layer1Hybrid']=df['first'].apply(lambda x:1 if 'Hybrid' in x else 0)
df['Layer1FlashStorage']=df['first'].apply(lambda x:1 if 'Flash Storage' in x else 0)

df['first']=df['first'].str.replace(r'\D','',regex=True)
#used to remove all non-digit characters from the values in the "first" column, where \D is for non-digit character



df['second']=new[1]
df['second']=df['second'].fillna('0')
df['second']=df['second'].str.strip()
#This is to make all None or missing value to be 0
df['Layer2HDD']=df['second'].apply(lambda x:1 if 'HDD' in x else 0)
df['Layer2SSD']=df['second'].apply(lambda x:1 if 'SSD' in x else 0)
df['Layer2Hybrid']=df['second'].apply(lambda x:1 if 'Hybrid' in x else 0)
df['Layer2FlashStorage']=df['second'].apply(lambda x:1 if 'Flash Storage' in x else 0)

df['second']=df['second'].str.replace(r'\D','',regex=True)




df['first']=df['first'].astype(int)
df['second']=df['second'].astype(int)
df['HDD']=df['first']*df['Layer1HDD']+df['second']*df['Layer2HDD']
df['SSD']=df['first']*df['Layer1SSD']+df['second']*df['Layer2SSD']
df['Hybrid']=df['first']*df['Layer1Hybrid']+df['second']*df['Layer2Hybrid']
df['FlashStorage']=df['first']*df['Layer1FlashStorage']+df['second']*df['Layer2FlashStorage']


df.drop(columns=['first','second','Layer1SSD','Layer1HDD','Layer1Hybrid','Layer1FlashStorage',
                 'Layer2SSD','Layer2HDD','Layer2Hybrid','Layer2FlashStorage'],inplace=True)
df.sample(5)
```

MODEL BUILDING

```python
X = df.drop(columns=['Price'])
y = np.log(df['Price'])
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import r2_score, mean_absolute_error

# Create a ColumnTransformer for one-hot encoding the categorical features
encoder = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])],
    remainder='passthrough'
)

# Fit and transform the training data
X_encoded=encoder.fit_transform(X)
X_train,X_test,y_train,y_test = train_test_split(X_encoded,y,test_size=0.2,random_state=42)
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression

# Create and fit the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Calculate R2 score and mean absolute error
r2_score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print('R2 score:', r2_score)
print('MAE:', mae)
```

Ridge Regression



```python
#If you choose run all option, you should run again the model building part in 2 cells above before run this.
#Create and fit the Ridge Regression
from sklearn.linear_model import Ridge
#Define the alpha value to test
alpha_values=[0.1,0.5,2,5,10,20,50,100]
r2_scores=[]
mae_scores=[]
for alpha in alpha_values:
    model=Ridge(alpha=alpha)
    model.fit(X_train,y_train)
    # Make predictions on the test data
    y_pred = model.predict(X_test)

    # Calculate R2 score and mean absolute error
    r2_scores.append(r2_score(y_test, y_pred))
    mae_scores.append(mean_absolute_error(y_test, y_pred))

#Visualize the scores:
plt.figure(figsize=(10,6))
plt.plot(alpha_values,r2_scores,'bo-',label='R2_score')
plt.plot(alpha_values,mae_scores,'bo-',label='MAE score')
plt.xlabel('Alpha')
plt.ylabel('Score')
plt.grid(True)
plt.title("Visualize the score with alpha changing")
plt.show()


```

```python
#Seem that the value alpha=5 or alpha=10 would be appropriate, because a small alpha will lead to the small impact of penalty term, which can lead to over
#fitting in the future, while a large alpha will have the penalty term dominates the cost function, and the coefficients are heavily penalized. 
#This can result in significant shrinkage of the coefficients, which may lead to underfitting. Hence I choose alpha=10 to work with this model.
```

```python
from sklearn.linear_model import Ridge
model=Ridge(alpha=10)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
# Calculate R2 score and mean absolute error
r2score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print('R2 score:', r2score)
print('MAE:', mae)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of values for n_neighbors
n_neighbors_values = [5, 10, 15, 20]

# Initialize empty lists to store the evaluation scores
r2_scores = []
mae_scores = []

# Iterate over each value of n_neighbors
for n in n_neighbors_values:
    # Create and fit the KNeighborsRegressor model
    model = KNeighborsRegressor(n_neighbors=n)
    model.fit(X_train, y_train)
    
    # Make predictions on the test data
    y_pred = model.predict(X_test)
    
    # Calculate R2 score and mean absolute error
    r2score = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    
    # Append the scores to the respective lists
    r2_scores.append(r2score)
    mae_scores.append(mae)

# Plotting the scores
plt.plot(n_neighbors_values, r2_scores, label='R2 Score')
plt.plot(n_neighbors_values, mae_scores, label='MAE')
plt.xlabel('n_neighbors')
plt.ylabel('Score')
plt.title('Evaluation Scores for different n_neighbors')
plt.legend()
plt.show()
```

Random Forest 

```python

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error




# Define the range of values for n_estimators
n_estimators_range = [1, 5, 10, 20, 50, 100]

# Initialize lists to store the performance metrics
r2_scores = []
mae_scores = []

# Iterate over different values of n_estimators
for n_estimators in n_estimators_range:
    # Create and fit the RandomForestRegressor
    model = RandomForestRegressor(n_estimators=n_estimators, random_state=42)
    model.fit(X_train, y_train)
    
    # Make predictions on the test data
    y_pred = model.predict(X_test)
    
    # Calculate the performance metrics
    r2score = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    
    # Append the scores to the lists
    r2_scores.append(r2score)
    mae_scores.append(mae)

# Plot the performance metrics
plt.figure(figsize=(10, 6))
plt.plot(n_estimators_range, r2_scores, label='R2 Score')
plt.plot(n_estimators_range, mae_scores, label='MAE')
plt.xlabel('Number of Trees (n_estimators)')
plt.ylabel('Performance')
plt.title('Performance vs. Number of Trees')
plt.legend()
plt.show()
```

Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score, mean_absolute_error




# Define a list of number of estimators to try
n_estimators = [1, 5, 10, 50, 100, 200]

# Initialize lists to store the accuracy scores
r2_scores = []
mae_scores = []

# Iterate over different numbers of estimators
for n in n_estimators:
    # Create and fit the gradient boosting regressor
    model = GradientBoostingRegressor(n_estimators=n, random_state=42)
    model.fit(X_train, y_train)
    
    # Make predictions on the test set
    y_pred = model.predict(X_test)
    
    # Calculate accuracy metrics
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    
    # Append scores to the respective lists
    r2_scores.append(r2)
    mae_scores.append(mae)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(n_estimators, r2_scores, label='R2 Score')
plt.plot(n_estimators, mae_scores, label='MAE')
plt.xlabel('Number of Estimators')
plt.ylabel('Accuracy')
plt.title('Accuracy of Gradient Boosting with Different Number of Estimators')
plt.legend()
plt.show()


#This method is started by making initial predictions using a simple model, then Fit a new model (e.g., a decision tree) to the residuals. 
# This new model will try to predict the residuals and capture the patterns that the previous model missed.

#Update the predictions: Adjust the predictions by adding the predictions from the new model, multiplied by a small learning rate. 
# The learning rate controls the contribution of each new model to the overall predictions.
#Therefore the accuracy increases when more iterators are used.
```

```python
#I will choose n_estimators=100 for prediction
model=GradientBoostingRegressor(n_estimators=100,random_state=42)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
r2score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print('R2 score:', r2score)
print('MAE:', mae)
```

```python

import matplotlib.pyplot as plt
from sklearn.ensemble import VotingRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score

# Define the individual regressor models
rf = RandomForestRegressor(n_estimators=50, bootstrap=True, random_state=42, max_samples=0.5, max_features=0.75, max_depth=10)
gbr = GradientBoostingRegressor(n_estimators=100, random_state=42, max_features=0.5)
xgb = XGBRegressor(n_estimators=25, learning_rate=0.3, max_depth=5)

# Create the VotingRegressor ensemble
model = VotingRegressor([('rf', rf), ('gbr', gbr), ('xgb', xgb)])

# Fit the ensemble model
model.fit(X_train, y_train)
rf.fit(X_train,y_train)
gbr.fit(X_train,y_train)
xgb.fit(X_train,y_train)

# Make predictions on the test data
y_pred_rf = rf.predict(X_test)
y_pred_gbr = gbr.predict(X_test)
y_pred_xgb = xgb.predict(X_test)

# Calculate the R2 score for each regressor
r2score_rf = r2_score(y_test, y_pred_rf)
r2score_gbr = r2_score(y_test, y_pred_gbr)
r2score_xgb = r2_score(y_test, y_pred_xgb)

# Create a bar plot of R2 scores for each regressor
regressors = ['Random Forest', 'Gradient Boosting', 'XGBoost']
r2_scores = [r2score_rf, r2score_gbr, r2score_xgb]

plt.bar(regressors, r2_scores)
plt.xlabel('Regressor')
plt.ylabel('R2 Score')
plt.title('R2 Score Comparison for Each Regressor')

# Display the detailed R2 scores as text on the plot
for i, v in enumerate(r2_scores):
    plt.text(i, v, str(round(v, 4)), ha='center', va='bottom')

plt.show()


```

Stacking Regressor


```python
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import LinearRegression

# Define the base regressors
base_regressors = [('rf', rf), ('gbr', gbr), ('xgb', xgb)]

# Initialize the stacking regressor with the base regressors and a final meta regressor
stacking_model = StackingRegressor(estimators=base_regressors, final_estimator=LinearRegression())

# Fit the stacking model on the training data
stacking_model.fit(X_train, y_train)

# Make predictions on the test data
y_pred_stacking = stacking_model.predict(X_test)

# Calculate the R2 score and mean absolute error
r2score_stacking = r2_score(y_test, y_pred_stacking)
mae_stacking = mean_absolute_error(y_test, y_pred_stacking)

print('Stacking R2 score:', r2score_stacking)
print('Stacking MAE:', mae_stacking)
```

In summary, the VotingRegressor uses a simple averaging scheme to combine the predictions, while the StackingRegressor learns a more sophisticated way to combine the predictions using a meta-regressor. The StackingRegressor has the potential to learn the optimal combination of the base regressors, which can lead to improved performance compared to the VotingRegressor
